```python
# Variance
# using numpy
import numpy as np
data = [2,4,4,5,6,7,8,9]
sample_variance = np.var(data, ddof= 1)   # ddof ----> data degree of freedom
pop_variance = np.var(data, ddof = 0)
print("the simple variance of the data is ",sample_variance)
print("the population variance of the data is ",pop_variance)
```

```python
# using pandas
import pandas as pd 
data = [2,4,4,5,6,7,8,9]
serise = pd.Series(data)
print(serise)
sample_variance = serise.var()
print("the sample variance is ",sample_variance)
pop_variance = serise.var(ddof=0)
print("the population variance is ",pop_variance)
```

```python
# using Scipy
from scipy import stats 
data = [2,4,4,5,6,7,8,9]
sample_variance = stats.tvar(data)
print("the sample variance is ",sample_variance)
pop_variance = stats.tvar(data,ddof=0)
print("the population variance is ",pop_variance)
```

```python
# Standard Deviation
# using numpy
import numpy as np
data = [60,65,70,75,80]
sample_std = np.std(data, ddof=1)
print("The sample variance of the data is ", sample_std)
population_std = np.std(data, ddof=0)
print("The population variance of the data is ", population_std)
```

```python
# Standard Pandas
import pandas as pd
data = [60,65,70,75,80]
serise = pd.Series(data)
sample_std = serise.std()
print("The sample Standard deviation of the data is ", sample_std)
population_std = serise.std(ddof=0)
print("The population Standard deviation of the data is ", population_std)
```

```python
# using Scipy
from scipy import stats
import math
data = [60,65,70,75,80]
sample_std = math.sqrt(stats.tvar(data))
print("This sample standard deviation is ",sample_std)
pop_std = math.sqrt(stats.tvar(data,ddof=0))
print("This population standard deviation is ",pop_std)
```

```python
# Data Representations
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
fig,axs = plt.subplots(3,2,figsize=(14,12))
fig.suptitle("Histograms with KDE of Various Distributions ", fontsize = 16)

data_normal = np.random.normal(loc=0 , scale = 1, size= 10000)
sns.histplot(data_normal, kde=True, ax= axs[0,0],color="blue",stat = 'density',bins=30,label="varianc")
axs[0,0].set_title("Normal Distribution")
axs[0,0].grid(True)

data_right_skewed = np.random.exponential(scale=1,size=1000)
data_left_skewed = -np.random.exponential(scale=1,size=1000)
sns.histplot(data_right_skewed,kde=True,ax=axs[0,1],color='red',stat='density',bins=30,label = "Right Skewed")
sns.histplot(data_left_skewed,kde=True,ax=axs[0,1],color='green',stat='density',bins=30,label = "left Skewed")
axs[0,1].set_title("Skewed Distribution")
axs[0,1].legend()
axs[0,1].grid(True)

data_uniform = np.random.uniform(low=0,high=1,size=1000)
sns.histplot(data_uniform,kde=True,ax=axs[1,0],color='purple',stat='density',bins=30,label = "uniform")
axs[1,0].set_title("Uniform Distribution")
axs[1,0].grid(True)

data_bimodal = np.concatenate([np.random.normal(loc=-2,scale=1,size=500),np.random.normal(loc=2,scale=1,size=500)])
sns.histplot(data_bimodal,kde=True,ax=axs[1,1],color='orange',stat='density',bins=30,label = "bimodel")
axs[1,1].set_title("Bimodal Distribution")
axs[1,1].grid(True)


data_maltimodel = np.concatenate([np.random.normal(loc=-6,scale=1.5,size=500),np.random.normal(loc=0,scale=1,size=500),np.random.normal(loc=6,scale=1.5,size=500)])
sns.histplot(data_maltimodel,kde=True,ax=axs[2,0],color='brown',stat='density',bins=30,label = "maltimodel")
axs[2,0].set_title("Multimodal Distribution")
axs[2,0].grid(True)
axs[2,1].axis('off')
plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```

```python
# T test & P test
import numpy as np
from scipy import stats

group_A =  np.random.normal(75,10,30)
group_B =  np.random.normal(80,12,30)

t_statistic , p_value = stats.ttest_ind(group_A,group_B)
print(f"T-statistic: {t_statistic}")
print(f"p-value: {p_value}")
alpha = 0.05
if p_value<= alpha:
    print("Regect the null hypothesis (Significant difference between groups)")
else:
    print("Fail to reject the null hypothesis (No significant difference between groups)")
```
