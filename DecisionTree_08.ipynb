```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
```

```python
data1 = sns.load_dataset('titanic')
data1.head()
```

```python
data = data1.drop(columns=['deck'])
# impute missing values of age, and fare using median
imputer = SimpleImputer(strategy='median')
data[['age','fare']] = imputer.fit_transform(data[['age','fare']])
imputer = SimpleImputer(strategy='most_frequent')
data[['embark_town','embarked']] = imputer.fit_transform(data[['embark_town','embarked']])
data.isnull().sum()
```

```python
# Encode the categorical and object variables using for loop and labelencoder
le = LabelEncoder()
for col in data.select_dtypes(include=['object','category']):
    data[col] = le.fit_transform(data[col])
data.head()
```

```python
x = data.drop(['survived','alive'],axis=1)
y = data['survived']
x_train , x_test, y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = DecisionTreeClassifier(criterion='entropy',)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
# evaluation model  

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# Save the decision tree classifier
from sklearn.tree import export_graphviz
export_graphviz(model,out_file='./saved_models/Decision_tree_01.dot',feature_names=x.columns,filled=True,rounded=True)

pickle.dump(model, open('./saved_models/DecisionTree_model.pkl', 'wb'))
```
